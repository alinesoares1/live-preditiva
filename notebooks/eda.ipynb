```python
import pandas as pd
```

```python
df = pd.read_csv("/Users/aline/Documents/live-preditiva/data/dataset_engajamento_clientes.csv", sep=",")
df.head()
```

```python
# Recência = quantos dias se passaram desde a última interação do cliente até uma data de referência
# garantindo que a coluna é datetime
df["data"] = pd.to_datetime(df["data"])

# data atual (hoje)
hoje = pd.Timestamp.today().normalize()

# cálculo da recência
df["recencia_dias"] = (hoje - df["data"]).dt.days
```
